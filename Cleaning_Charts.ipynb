```python
import pandas as pd 
import matplotlib.pyplot as plt
women_degrees=pd.read_csv("percent-bachelors-degrees-women-usa.csv")

women_degrees.head()
```

```python
### Improving astheytics in pandas plots

women_degrees["Men_bio"]=100-women_degrees['Biology']
ax=women_degrees.plot('Year',y=['Biology','Men_bio'],legend=False,title="Biology",ylim=(0,100),xticks=[1970,1980,1990,2000,2010])
ax.tick_params(bottom='off',left='off',right='off',top='off')
ax.spines['top'].set_visible(False) 
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False) 
```

```python
### Improving astheytics in matplotlib
fig,ax1=plt.subplots()
ax1.plot(women_degrees['Year'],women_degrees['Biology'],color='blue',label='Women')
ax1.plot(women_degrees['Year'],100-women_degrees['Biology'],color='green',label='Men')
ax1.tick_params(bottom='off',left='off',right='off',top='off')
ax1.set(title="Biology")
ax1.spines['top'].set_visible(False) 
ax1.spines['right'].set_visible(False)
ax1.spines['bottom'].set_visible(False)
ax1.spines['left'].set_visible(False) 
ax1.legend(loc='upper right') 

plt.show()

```

```python
fig=plt.figure(figsize=(12,12))
major_cats = ['Biology', 'Computer Science', 'Engineering', 'Math and Statistics']
for i in major_cats:
    ax=fig.add_subplot(2,2,1+major_cats.index(i))
    ax.plot(women_degrees['Year'],women_degrees[i],color='blue',label='Women')
    ax.plot(women_degrees['Year'],100-women_degrees[i],color='green',label='Men')
    ax.spines['top'].set_visible(False) 
    ax.spines['right'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.tick_params(bottom='off',left='off',right='off',top='off')
    ax.set(title=i)
    ax.set_ylim(0,100)
   
    if major_cats.index(i)==3:
        ax.legend(loc='upper right') 
plt.show()
        
    




```

```python
fig=plt.figure(figsize=(12,12))
major_cats = ['Biology', 'Computer Science', 'Engineering', 'Math and Statistics']
men_color=(255/255,128/255,14/255)
women_color=(0/255,107/255,164/255)
for i in major_cats:
    ax=fig.add_subplot(2,2,1+major_cats.index(i))
    ax.plot(women_degrees['Year'],women_degrees[i],color=women_color,label='Women',linewidth=3)
    ax.plot(women_degrees['Year'],100-women_degrees[i],color=men_color,label='Men',linewidth=3)
    ax.spines['top'].set_visible(False) 
    ax.spines['right'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.tick_params(bottom='off',left='off',right='off',top='off')
    ax.set(title=i)
    ax.set_ylim(0,100)
   
    if major_cats.index(i)==3:
        ax.legend(loc='upper right')
plt.show()

```

```python
##### Adding Anootation to the graphs 

stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']
fig = plt.figure(figsize=(18, 3))
cb_orange=(255/255,128/255,14/255)
cb_dark_blue=(0/255,107/255,164/255)
for sp in range(0,6):
    ax = fig.add_subplot(1,6,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[sp])
    ax.tick_params(bottom="off", top="off", left="off", right="off")
    if sp==0:
        ax.text(2005,87,"Men")
        ax.text(2002,8,"Women")
    elif sp==5:
        ax.text(2005,62,"Men")
        ax.text(2001,35,"Women")
        
plt.show()
```
